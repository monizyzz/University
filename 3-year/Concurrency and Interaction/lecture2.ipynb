```python
import pennylane as qml
import numpy as np
```

```python
n_qubits = 1

dev = qml.device("default.qubit", wires=n_qubits)
```

```python
n_qubits = 2

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev) # bind
def circuito():
        
    qml.Hadamard(wires=0) # 1ยบ qubit
    qml.Hadamard(wires=1) # 2ยบ qubit
    
    return qml.probs(wires=[0,1])
```

```python
circuito()
```

```python
qml.draw_mpl(circuito)()
```

```python
n_qubits = 2

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev) # bind
def circuito():
    
    #qml.PauliX(wires=[0])
    qml.Hadamard(wires=[0])
    
    qml.CNOT(wires=[0,1])
    
    return qml.probs(wires=[0,1])

circuito()
```

```python
qml.draw_mpl(circuito)()
```
